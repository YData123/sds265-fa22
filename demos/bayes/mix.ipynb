## Demo code for mixtures

In this little notebook we'll play around with mixture distributions. This will help to solidify concepts, which will be important as we gear up for Bayesian modeling, topic models, and other good things coming up!

First, we'll generate mixtures of two Gaussians.


```python
import numpy as np
from scipy.stats import bernoulli
from scipy.stats import norm
import seaborn as sbn

n = 10000
eta = .5
z = bernoulli.rvs(eta, size=n)
x1 = norm.rvs(loc=-10, scale=1, size=n)
x2 = norm.rvs(loc=10, scale=10, size=n)
x = z*x1 + (1-z)*x2

_ = sbn.displot(x, kind='kde')
```

```python
sbn.displot(x2)
```

Now let's mix together three Gaussians. To do this we'll "roll a three-sided die" to determine the latent variable.


```python
from scipy.stats import multinomial
n = 10000
eta = [.3, .5, .2]
sigma = 3
z = multinomial.rvs(p=eta, n=1, size=n)

x0 = norm.rvs(loc=-10, scale=sigma, size=n)
x1 = norm.rvs(loc=0, scale=sigma, size=n)
x2 = norm.rvs(loc=10, scale=2*sigma, size=n)

x = z[:,0]*x0 + z[:,1]*x1 + z[:,2]*x2

_ = sbn.displot(x, kind='kde')
```
